```python
from collections import defaultdict
import math

# Initialize dictionary to hold categorized data
data_by_date_range = defaultdict(list)

# Collect all dates from the filtered data to determine the min and max
dates = [int(details['date']) for filename, details in filtered_data_by_filename.items() if details['date'].isdigit()]

if not dates:
    print("No valid date data found in the filtered dictionary.")
    exit()

# Determine the range of dates and create 5 equal intervals
min_year = min(dates)
max_year = max(dates)
interval_length = math.ceil((max_year - min_year) / 5)

# Function to assign date ranges
def determine_date_range(year):
    if year < min_year:
        return f"Before {min_year}"
    elif year > max_year:
        return f"After {max_year}"
    range_index = (year - min_year) // interval_length
    range_start = min_year + range_index * interval_length
    range_end = range_start + interval_length - 1
    # Adjust the last interval to not exceed the max_year
    if range_end > max_year:
        range_end = max_year
    return f"{range_start}-{range_end}"

# Categorize entries by date range
for filename, details in filtered_data_by_filename.items():
    try:
        year = int(details['date'])
        date_range = determine_date_range(year)
        file_info = (filename, details['artist'], details['title'], details['style'], details['genre'], details['date'])
        data_by_date_range[date_range].append(file_info)
    except ValueError:
        continue  # Skip entries with invalid dates

# Output the data categorized by date range
for range_key, files in data_by_date_range.items():
    print(f"{range_key}: {len(files)} files")
```

Creating custom Data Generator

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Load your data
data = pd.read_csv('genre_classification.csv')

# Optionally shuffle the DataFrame if it's not already randomized
data = data.sample(frac=1).reset_index(drop=True)

# Split data into training and testing sets
train_data, test_data = train_test_split(data, test_size=0.15, random_state=42)  # 15% of the data is used for testing

# Separate filenames and labels for training and testing
train_filenames = train_data['filename'].values
train_labels = train_data['genre'].values

test_filenames = test_data['filename'].values
test_labels = test_data['genre'].values

# Encode labels
label_encoder = LabelEncoder()
train_labels_encoded = label_encoder.fit_transform(train_labels)
test_labels_encoded = label_encoder.transform(test_labels)  # Use the same encoder to ensure consistency
```

```python
import pandas as pd
import numpy as np
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from sklearn.preprocessing import LabelEncoder


def image_generator(filenames, labels, batch_size, img_directory, target_size=(64, 64)):
    while True:  # Loop indefinitely
        for start in range(0, len(filenames), batch_size):
            end = min(start + batch_size, len(filenames))
            batch_filenames = filenames[start:end]
            batch_labels = labels[start:end]

            # Load images and process them
            images = [img_to_array(load_img(f"{img_directory}/{img_name}", target_size=target_size)) for img_name in batch_filenames]
            images = np.array(images) / 255.0  # Normalize the images

            # Convert labels to one-hot encoded
            batch_labels = to_categorical(batch_labels, num_classes=len(label_encoder.classes_))

            yield images, batch_labels

# Create generators
batch_size = 256
img_directory = 'Training Images - 256'
train_generator = image_generator(train_filenames, train_labels_encoded, batch_size, img_directory)
test_generator = image_generator(test_filenames, test_labels_encoded, batch_size, img_directory)
```

Configuring Model (CNN in Keras)

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

# Build a simple CNN model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(64, 64, 3)),
    MaxPooling2D(2, 2),
    Conv2D(64, (3, 3), activation='relu'),
    # Removed one MaxPooling2D to preserve feature map size
    Flatten(),
    Dense(100, activation='relu'),
    Dropout(0.3),  # Adjusted dropout rate
    Dense(len(label_encoder.classes_), activation='softmax')
])
```

```python
import csv

# Define the CSV file path
csv_file_path = 'date_classification.csv'

# Open the CSV file and count the number of rows
with open(csv_file_path, 'r', newline='') as csvfile:
    reader = csv.reader(csvfile)
    row_count = sum(1 for row in reader)

print(f"Number of rows in the CSV file: {row_count}")
```

```python
# import tools
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Load your data
data = pd.read_csv('date_classification.csv')

# Optionally shuffle the DataFrame if it's not already randomized
data = data.sample(frac=1).reset_index(drop=True)

# Split data into training and testing sets
train_data, test_data = train_test_split(data, test_size=0.15, random_state=42)  # 15% of the data is used for testing

# Separate filenames and labels for training and testing
train_filenames = train_data['filename'].values
train_labels = train_data['date range'].values

test_filenames = test_data['filename'].values
test_labels = test_data['date range'].values

# Encode labels
label_encoder = LabelEncoder()
train_labels_encoded = label_encoder.fit_transform(train_labels)
test_labels_encoded = label_encoder.transform(test_labels)  # Use the same encoder to ensure consistency
```

```python
import pandas as pd
import numpy as np
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from sklearn.preprocessing import LabelEncoder

def image_generator(filenames, labels, batch_size, img_directory, target_size=(64, 64)):
    while True:  # Loop indefinitely
        for start in range(0, len(filenames), batch_size):
            end = min(start + batch_size, len(filenames))
            batch_filenames = filenames[start:end]
            batch_labels = labels[start:end]

            # Load images and process them
            images = [img_to_array(load_img(f"{img_directory}/{img_name}", target_size=target_size)) for img_name in batch_filenames]
            images = np.array(images) / 255.0  # Normalize the images

            # Convert labels to one-hot encoded
            batch_labels = to_categorical(batch_labels, num_classes=len(label_encoder.classes_))

            yield images, batch_labels

# Create generators
batch_size = 64
img_directory = 'Training Images New'
train_generator = image_generator(train_filenames, train_labels_encoded, batch_size, img_directory)
test_generator = image_generator(test_filenames, test_labels_encoded, batch_size, img_directory)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

# Build a simple CNN model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(64, 64, 3)),
    MaxPooling2D(2, 2),
    Conv2D(64, (3, 3), activation='relu'),
    # Removed one MaxPooling2D to preserve feature map size
    Flatten(),
    Dense(100, activation='relu'),
    Dropout(0.3),  # Adjusted dropout rate
    Dense(len(label_encoder.classes_), activation='softmax')
])
```

```python
model.summary()
```

```python
# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(train_generator, steps_per_epoch=len(train_filenames) // batch_size, epochs=30)
```

```python
model.save('date_classification.h5')  # Saves the model in HDF5 format
```

```python
test_loss, test_accuracy = model.evaluate(test_generator, steps=len(test_filenames) // batch_size)
```
